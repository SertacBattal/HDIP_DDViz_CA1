## Part 1

### • What are the top 5 games by global sales?  

```python
sorted_games = vgsales_df.sort_values(by='Global_Sales', ascending = False)

top_5_games = sorted_games.head(5)

top_5_games[['Name', 'Global_Sales']]
```

```python
top_5_games[['Name', 'Global_Sales', 'Platform']]
```

```python
import matplotlib.pyplot as plt

colors = ['cornflowerblue', 'red', 'cornflowerblue', 'cornflowerblue', 'gold']
plt.figure(figsize=(8, 6))

bars = plt.bar(top_5_games['Name'], top_5_games['Global_Sales'], color = colors,edgecolor = "black", linewidth = 1)

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 3, round(yval, 2), ha='center', va='top', fontsize=10)

plt.xlabel('Game Name')
plt.ylabel('Global Sales')
plt.title('Top 5 Games by Global Sales')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Figure 1
plt.text(0, -0.5, 'Figure 1', ha='left', va='bottom', transform=plt.gca().transAxes, fontsize=12, fontstyle='italic')

# Grafiği gösterme
plt.show()
```

### • Is there a correlation between the “na_sales” and “jp_sales” for the years 2010-2014? 

```python
# Selecting data from the years 2010-2014
filtered_df = vgsales_df[(vgsales_df['Year_of_Release'] >= 2010) & (vgsales_df['Year_of_Release'] <= 2014)]

# Calculate the correlation between the "na_sales" and "jp_sales" columns
correlation = filtered_df['NA_Sales'].corr(filtered_df['JP_Sales'])
correlation
```

Correlation coefficient is a statistical measure that measures the strength and direction of the relationship between two variables. It takes a value between -1 and 1.<br>
Coefficient value:<br>
The closer it gets to 1, the stronger and more positive the relationship between variables is. That is, as one variable increases, the other variable also increases.<br>
As it approaches -1, the relationship between variables is stronger and more negative. That is, as one variable increases, the other variable decreases.<br>
Closer to 0, the relationship between variables is weak or non-existent.

The correlation coefficient between variables such as "na_sales" and "jp_sales" is 0.26, the relationship between the variables is weak. So we cannot say that as North American sales increase, Japan sales will generally increase as well.

```python
plt.figure(figsize=(8, 6))
plt.scatter(filtered_df['NA_Sales'], filtered_df['JP_Sales'],c=filtered_df['Year_of_Release'], cmap='viridis', alpha=0.7)

plt.xlabel('NA Sales')
plt.ylabel('JP Sales')
plt.title('Correlation between NA Sales and JP Sales for 2010-2014')
plt.colorbar(label='Years')
plt.tight_layout()

# Figure 2
plt.text(0, -0.2, 'Figure 2', ha='left', va='bottom', transform=plt.gca().transAxes, fontsize=12, fontstyle='italic')

plt.show()
```

### • Do older games (2005 and earlier) have a higher MEAN “eu_sales” than newer games (after 2005)?  

```python
np.sort(vgsales_df['Year_of_Release'].unique())
```

```python
older_games_mean_sales = vgsales_df[vgsales_df['Year_of_Release'] <= 2005]['EU_Sales'].mean()
newer_games_mean_sales = vgsales_df[vgsales_df['Year_of_Release'] > 2005]['EU_Sales'].mean()

print("Mean EU Sales for 2005 and earlier: {:.3f}".format(older_games_mean_sales))
print("Mean EU Sales for After 2005:       {:.3f}".format(newer_games_mean_sales))
```

Yukarida gordugumuz gibi 2005 ve oncesi(25 yillik deger(ilk oyun yili veri setinde 1980)) EU satislarinin mean degeri, 2005 yilindan sonra ki EU satislari (15 yillik deger(son oyun yili veri setinde 2020)) cok az degerde yuksek, hatta esit denilebilir.

```python
# Verileri çubuk grafikle görselleştirelim
plt.figure(figsize=(8, 6))
sns.barplot(x=['Older Games', 'Newer Games'], y=[older_games_mean_sales, 
                                                 newer_games_mean_sales], palette='pastel')

plt.axhline(y=older_games_mean_sales, color='red', linestyle='--')
plt.axhline(y=newer_games_mean_sales, color='red', linestyle='--')

plt.xlabel('Game Category')
plt.ylabel('Mean EU Sales')
plt.title('Mean EU Sales Comparison between Older and Newer Games')
plt.tight_layout()

# Figure 4
plt.text(0, -0.2, 'Figure 4', ha='left', va='bottom', transform=plt.gca().transAxes, fontsize=12, fontstyle='italic')

# Grafiği gösterelim
plt.show()
```
